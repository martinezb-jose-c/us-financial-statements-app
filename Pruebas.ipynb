```python
# ✅ FUNCIÓN CORREGIDA - COPIA ESTO EN UNA NUEVA CELDA

def get_statements_by_primary_concept(filing, primary_concept):
    """
    Obtiene los statements filtrados por primary_concept.

    Args:
        filing: Objeto EntityFiling
        primary_concept: String del concepto a buscar (ej: 'us-gaap_DebtDisclosureAbstract')

    Returns:
        Dict con los statements que coinciden
    """
    # Obtener statements - filing YA TIENE el método directamente
    statements_dict = filing.get_statements_by_category()

    # Filtrar por primary_concept
    filtered = {}
    for category_key, category_items in statements_dict.items():
        # category_items es una lista de statements
        if isinstance(category_items, list):
            for item in category_items:
                if isinstance(item, dict):
                    if item.get('primary_concept') == primary_concept:
                        # Guardar en dict con key único
                        key = f"{category_key}_{item.get('index')}"
                        filtered[key] = item

    return filtered


# ✅ PRUEBA: Obtener todos los Debt Disclosures
debt_statements = get_statements_by_primary_concept(
    filing,
    'us-gaap_DebtDisclosureAbstract'
)

print(f"✅ Encontrados {len(debt_statements)} roles de Deuda\n")
for role_key, info in debt_statements.items():
    print(f"📋 {info.get('role_name')}")
    print(f"   Definición: {info.get('definition')}")
    print(f"   Elementos: {info.get('element_count')}")
    print(f"   URL: {info.get('role')}")
    print()
```
